```python
import numpy as np
import marshal, json
import matplotlib.pyplot as plt
import json, math
from IPython.display import set_matplotlib_formats
set_matplotlib_formats('svg')


plt.rcParams['figure.dpi'] = 300
plt.rcParams['savefig.dpi'] = 300

def get_dist(q):
    with open(f'../res/v24_parsed/rdf_n300_i120_q{q}_r2_aoi_dist_r1.dat', 'rb') as f:
        return marshal.load(f)
        



for q in [100, 150, 200, 250]:
    t_horizon = 1000

    dist = get_dist(q)
    p = np.array(dist['p'])
    t = np.array(dist['t'])
    plt.plot(t, np.cumsum(p))
    plt.axis([0, t_horizon, 0, 1])
    plt.show()
    print(np.sum(p[t <= t_horizon]) * 100)
```
