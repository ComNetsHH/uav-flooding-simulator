```python
fig, ax = plt.subplots()

i = 0.2

q_list= [0.6, 0.7, 0.8, 0.9, 1.0, 1.1, 1.2, 1.3, 1.4, 1.5]

h = np.arange(0, 20)

for q in q_list:
    ax.plot(h, i * q**h, color="#b50600")

for q in q_list:
    ax.plot(h, i * (h+1)**q, color="#f1a92f")

for q in q_list:
    ax.plot(h, i * q * (h+1), color="#005e8d")

ax.set_ylim(0, 20)
ax.set_xlim(0, 20)
ax.set_xticks(np.arange(0,20,2))

ax.set_ylabel("Forwarding Interval at Hop h [s]")
ax.set_xlabel("Hop h")

plt.title("Behavior of Different Decay-Functions $i=200$ms")
ax.plot(0,0, color="#b50600", label="Exponential Decay")
ax.plot(0,0, color="#f1a92f", label="Polynomial Decay")
ax.plot(0,0, color="#005e8d", label="Linear Decay")


ax.spines.top.set_visible(False)
ax.spines.top.set_visible(False)

leg = plt.legend(fancybox=False, framealpha=1.0)
leg.get_frame().set_edgecolor('#333')

fig.savefig(f"../figures/decay_functions.pdf", dpi=500, bbox_inches='tight', pad_inches=0.01)
fig.savefig(f"../figures/decay_functions.png", dpi=500, bbox_inches='tight', pad_inches=0.01)
```
