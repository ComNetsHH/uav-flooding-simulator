```python
### IMPORTS
import numpy as np
import matplotlib.pyplot as plt
import matplotlib_inline.backend_inline

matplotlib_inline.backend_inline.set_matplotlib_formats('svg')
```

```python
### DATA
time = np.array([2035, 2040, 2045, 2050])
A = 12012

# Table 8
parcel_delivery_low = np.array([33904, 35445, 36987, 39041]) / A
parcel_delivery_mid = np.array([39041, 42637, 46747, 51370]) / A
parcel_delivery_high = np.array([51885, 61644, 73923, 88356]) / A

# Table 11
food_delivery_low = np.array([21303, 22279, 23300, 24367]) / A
food_delivery_mid = np.array([24555, 26846, 29351, 32089]) / A
food_delivery_high = np.array([32504, 38791, 46295, 55250]) / A
```

```python
### VISUALIZATION

fig, ax = plt.subplots() 

ax.plot(time, food_delivery_low + parcel_delivery_low)
ax.plot(time, food_delivery_mid + parcel_delivery_mid)
ax.plot(time, food_delivery_high + parcel_delivery_high)

ax.grid()
plt.show()

print(food_delivery_low + parcel_delivery_low)
```
