```python
### IMPORTS
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib_inline.backend_inline

matplotlib_inline.backend_inline.set_matplotlib_formats('svg')
```

```python
df = pd.read_csv('../res/v30/course_rdf_n100_i60_q100_r0.csv')
fig, ax = plt.subplots()
for i in range(300):
    trajectory = df[df['nodeId'] == i]
    ax.plot(trajectory['pos_x'], trajectory['pos_y'])


fig.savefig(f"../figures/trajectories.png", dpi=500, bbox_inches='tight', pad_inches=0.01)
plt.show()
```
